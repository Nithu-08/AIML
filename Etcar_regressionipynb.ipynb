```python
def remove_outliers(df, columns):
    for column in columns:
        Q1 = np.percentile(df[column], 25, interpolation='midpoint')
        Q3 = np.percentile(df[column], 75, interpolation='midpoint')
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df

m1 = remove_outliers(m, ['Priceofthecar', 'ResaleValue' , 'Width', 'Length','FuelEfficiency', 'Salesinthousands'])
m1.head()
```

```python
# prompt: perform regression with sales as dependent and price, engine size,horsepower, width, fuel as independent

import statsmodels.api as sm

# Define the dependent and independent variables
dependent_variable = m3['Salesinthousands']
independent_vars = m3[['Priceofthecar', 'Enginesize', 'Horsepower', 'Width', 'FuelEfficiency']]

# Add a constant term to the independent variables
independent_vars_with_constant = sm.add_constant(independent_vars)

# Create the model
model = sm.OLS(dependent_variable, independent_vars_with_constant)

# Fit the model
model_fit = model.fit()

# Print the model summary
print(model_fit.summary())
```
