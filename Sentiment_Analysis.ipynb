```python
# prompt: perform sentiment analysis on the above data

from textblob import TextBlob

def get_sentiment(text):
  """
  This function takes a string as input and returns its sentiment as a float between -1 and 1.
  """
  analysis = TextBlob(text)
  return analysis.sentiment.polarity

p['Sentiment'] = p['Review'].apply(get_sentiment)

# Print the head of the DataFrame with the new 'Sentiment' column
p.head()
```

```python
# prompt: divide it into positive , negative and neutral

def categorize_sentiment(sentiment):
  """
  This function takes a sentiment score as input and returns a string representing the sentiment category.
  """
  if sentiment > 0:
    return "Positive"
  elif sentiment < 0:
    return "Negative"
  else:
    return "Neutral"

p['Sentiment_Category'] = p['Sentiment'].apply(categorize_sentiment)

# Print the head of the DataFrame with the new 'Sentiment_Category' column
p.head(7)
```

```python
# prompt: perform sentiment analysis for Likes A, Dislikes A, Likes B, Dislikes B, Likes C, Dislikes C into positive negative and neutral

# Assuming 'Likes A', 'Dislikes A', ... columns contain text reviews
!pip install textblob
from textblob import TextBlob
def get_sentiment(text):
  """
  This function takes a string as input and returns its sentiment as a float between -1 and 1.
  """
  # Handle potential non-string values
  if not isinstance(text, str):
    return None  # Or handle it in a way that suits your data
  analysis = TextBlob(text)
  return analysis.sentiment.polarity

def categorize_sentiment(sentiment):
  """
  This function takes a sentiment score as input and returns a string representing the sentiment category.
  """
  if sentiment > 0:
    return "Positive"
  elif sentiment < 0:
    return "Negative"
  else:
    return "Neutral"

# Perform sentiment analysis on each column
for col in ['Likes A', 'Dislikes A', 'Likes B', 'Dislikes B', 'Likes c', 'Dislikes C']:
  # Convert the column to string type before applying sentiment analysis
  g[col] = g[col].astype(str)  # Convert column to string type
  g[col + '_Sentiment'] = g[col].apply(get_sentiment)
  g[col + '_Sentiment_Category'] = g[col + '_Sentiment'].apply(categorize_sentiment)

# Print the head of the DataFrame with the new sentiment columns
g.head()
```
